Se importan las librerías necesarias para el desarrollo de la práctica.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Se carga la imagen desde una ruta específica utilizando la función cv2.imread(). Se emplea cv2.IMREAD_GRAYSCALE para convertir la imagen directamente a escala de grises, lo que significa que cada píxel de la imagen tendrá un valor de intensidad entre 0 y 255.

```python
# Cargar la imagen en escala de grises
imagen = cv2.imread("imagen P4.jpeg", cv2.IMREAD_GRAYSCALE)
```

Después se aplica la ecualización de histograma mediante la función cv2.equalizeHist(). Este proceso consiste en redistribuir los valores de intensidad de los píxeles de la imagen para abarcar un rango más amplio, lo que permite mejorar el contraste.

```python
# Aplicar ecualización de histograma
imagen_ecualizada = cv2.equalizeHist(imagen)
```

Posteriormente se calculan los histogramas tanto de la imagen original como de la imagen ecualizada utilizando la función cv2.calcHist(). El histograma representa la distribución de los niveles de intensidad de la imagen, donde el eje horizontal corresponde a los valores de gris (de 0 a 255) y el eje vertical indica la frecuencia de aparición de cada valor.

```python
# Calcular histogramas antes y después
hist_original = cv2.calcHist([imagen], [0], None, [256], [0,256])
hist_ecualizada = cv2.calcHist([imagen_ecualizada], [0], None, [256], [0,256])
```

Finalmente se realiza una visualización de la imagen original y la imagen ecualizada junto con sus respectivos histogramas para realizar una comparación entre estos.

```python
# Imagen Original
plt.subplot(2,2,1)
plt.imshow(imagen, cmap="gray")
plt.title("Imagen Original")
plt.axis("off")

# Histograma Original
plt.subplot(2,2,2)
plt.plot(hist_original, color="black")
plt.title("Histograma Original")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

# Imagen Ecualizada
plt.subplot(2,2,3)
plt.imshow(imagen_ecualizada, cmap="gray")
plt.title("Imagen Ecualizada")
plt.axis("off")

# Histograma Ecualizado
plt.subplot(2,2,4)
plt.plot(hist_ecualizada, color="black")
plt.title("Histograma Ecualizado")
plt.xlabel("Intensidad")
plt.ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```
