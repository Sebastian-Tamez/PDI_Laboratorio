```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Primero cargamos la imagen desde el archivo.
# OpenCV la lee como una matriz numérica.
# Ojo: OpenCV usa formato BGR (Azul, Verde, Rojo), no RGB.
imagen = cv2.imread('flores-verano-jardin.jpg')

# Siempre verificamos que la imagen realmente se haya cargado.
# Si la ruta está mal, imread devuelve None.
if imagen is None:
    print("Error: No se pudo cargar la imagen.")
else:

    # Mostramos las dimensiones originales.
    # shape devuelve (alto, ancho, canales).
    # Si tiene 3 canales, es una imagen a color.
    print("Dimensiones originales (alto, ancho, canales):")
    print(imagen.shape)

    # Convertimos de BGR a RGB.
    # Esto es necesario porque matplotlib espera el orden RGB.
    imagen_rgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)

    # Mostramos la imagen ya convertida.
    # Si no hacemos la conversión anterior,
    # los colores se verán alterados.
    plt.figure(figsize=(6,6))
    plt.imshow(imagen_rgb)
    plt.title("Imagen en RGB")
    plt.axis('off')
    plt.show()

    # Ahora convertimos la imagen en una matriz de píxeles.
    # reshape(-1, 3) transforma la imagen de:
    # (alto, ancho, 3)
    # a
    # (numero_total_pixeles, 3)
    # donde cada fila representa un píxel con sus valores R, G y B.
    pixeles = imagen_rgb.reshape((-1, 3))

    print("Dimensiones como matriz de píxeles (N, 3):")
    print(pixeles.shape)

    # Mostramos los primeros 5 píxeles.
    # Cada fila es un vector [R, G, B].
    print("Primeros 5 vectores RGB:")
    print(pixeles[:5])
```

```python
print("Canal R promedio:", np.mean(imagen_rgb[:,:,0]))
print("Canal G promedio:", np.mean(imagen_rgb[:,:,1]))
print("Canal B promedio:", np.mean(imagen_rgb[:,:,2]))
```
